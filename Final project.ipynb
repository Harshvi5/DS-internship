```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('exercise.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='pulse', hue='diet', kde=True, bins=30)
plt.title('Distribution of Resting Heart Rates by Diet Type')
plt.xlabel('Resting Heart Rate')
plt.ylabel('Frequency')
plt.legend(title='Diet Type')
plt.grid(True)
plt.show();
```

```python
#3]
plt.figure(figsize=(12, 6))

for diet_type in df['diet'].unique():
    diet_df = df[df['diet'] == diet_type]
    sns.histplot(data=diet_df, x='pulse', hue='kind', kde=True, bins=30, label=diet_type)

plt.title('Distribution of Heart Rates Before and After Exercise by Diet Type')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.legend(title='Diet Type')

plt.tight_layout()
plt.show()

```

```python
#4.Calculate the average heart rate by diet type and activity level.

average_heart_rate = df.groupby(['diet', 'kind'])['pulse'].mean()
print(average_heart_rate)
```

```python
#5.Calculate average heart rate by diet type and activity level.

average_heart_rate = df.groupby(['diet', 'kind'])['pulse'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=average_heart_rate, x='diet', y='pulse', hue='kind')
plt.title('Average Heart Rate by Diet Type and Activity Level')
plt.xlabel('Diet Type')
plt.ylabel('Average Heart Rate')
plt.legend(title='Activity Level')
plt.show()
```

```python
#7. Plot the distribution of heart rates before and after exercise by gender.

plt.figure(figsize=(12, 6))

# Before Exercise
plt.subplot(1, 2, 1)
sns.boxplot(x='diet', y='pulse', hue='time', data=df[df['time'] == '1 min'])
plt.title('Heart Rate Before Exercise')

# After Exercise
plt.subplot(1, 2, 2)
sns.boxplot(x='diet', y='pulse', hue='time', data=df[df['time'] == '30 min'])
plt.title('Heart Rate After Exercise')

plt.tight_layout()
plt.show()
```

```python
#8. Calculate the median heart rate for each activity level.
median_heart_rate_by_activity = df.groupby('kind')['pulse'].median()

print(median_heart_rate_by_activity)
```
